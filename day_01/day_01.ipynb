```python
# 
# 13 

#  sans-serif
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei']  # Windows /

#  '-' 
plt.rcParams['axes.unicode_minus'] = False

# 1 x y
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]
y_beijing = [random.uniform(1, 3) for i in x]

# 2
# plt.figure(figsize=(20, 8), dpi=80)
figure, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 8), dpi=80)

# 3
axes[0].plot(x, y_shanghai, color="r", linestyle="-.", label="")
axes[1].plot(x, y_beijing, color="b", label="")

# 
axes[0].legend()
axes[1].legend()

# xy
# x
x_label = ["11{}".format(i) for i in x]
axes[0].set_xticks(x[::5])
axes[0].set_xticklabels(x_label[::5])
axes[0].set_yticks(range(0, 40, 5))
axes[1].set_xticks(x[::5])
axes[1].set_xticklabels(x_label[::5])
axes[1].set_yticks(range(0, 40, 5))

# 
axes[0].grid(linestyle="--", alpha=0.5)
axes[1].grid(linestyle="--", alpha=0.5)

# 
axes[0].set_xlabel("")
axes[0].set_ylabel("")
axes[0].set_title("1112")
axes[1].set_xlabel("")
axes[1].set_ylabel("")
axes[1].set_title("1112")

# 4
plt.show()
```
